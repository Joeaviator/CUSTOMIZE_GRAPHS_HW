```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('FuelConsumption.csv')
df.head()
```

```python
sns.set_theme(style="darkgrid")
sns.countplot(x='FUELTYPE', data=df, palette='rainbow')
plt.title('Fuel Type Distribution')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.countplot(x='FUELTYPE', data=df, palette='rainbow')
plt.title('Fuel Type Distribution')
sns.despine()
plt.show()
```

```python
sns.set_theme(style="white")
sns.set_context('paper')
top_10_makes = df['MAKE'].value_counts().nlargest(10).index
df_top_10 = df[df['MAKE'].isin(top_10_makes)]
sns.countplot(x='MAKE', data=df_top_10, palette='turbo', order=top_10_makes, hue='MAKE', legend=False)
plt.title('Top 10 MAKE Distribution')
sns.despine()
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_theme(style="white")
sns.set_context('poster')
top_10_makes = df['MAKE'].value_counts().nlargest(10).index
df_top_10 = df[df['MAKE'].isin(top_10_makes)]
sns.countplot(x='MAKE', data=df_top_10, color='turquoise', order=top_10_makes)
plt.title('Top 10 MAKE Distribution')
sns.despine()
plt.xticks(rotation=90)
plt.show()
```
